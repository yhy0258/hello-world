```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
import numpy as np
x=np.linspace(0,10,30)
y=np.sin(x)
```

```python
plt.plot(x,y,'ok')
```

```python
plt.plot(x,y,'-ok')
```

```python
plt.plot(x,y,'-p',color='green',
        markersize=15,linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2,
        )
plt.ylim(-1.2,1.2)
```

```python
plt.scatter(x,y,marker='o')
```

```python
rng=np.random.RandomState(0)
xr=rng.randn(100)
yr=rng.randn(100)
colors=rng.rand(100)
sizes=1000*rng.rand(100)
```

```python
plt.scatter(xr,yr,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar()
```

scatter单独性高，效率低
