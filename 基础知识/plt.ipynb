```python
import matplotlib.pyplot as plt
#换成R的风格
plt.style.use('ggplot')
#设置插图，使笔记本可见
%matplotlib inline
import numpy as np
```

```python
import os
import pandas as pd
PATH = r'E:\machine_learning\DATA\iris'
df = pd.read_csv(PATH + '\iris.data', names=['sepal length', 'sepal width','petal length', 'petal width', 'class'])
df.head()
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['petal width'], color='black');
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Width', fontsize=12)
plt.title('Iris Petal Width', fontsize=14, y=1.01)

#第一行创建了宽度为 6 英寸和高度为 4 英寸的一个插图。
#然后，我们通过调用.hist()并传入数据，依照 iris 数据框绘制了花瓣宽度的直方图。这里还将直方图中柱子的颜色设置为 black（黑色）。
#接下来的两行分别在 y 轴和 x 轴上放置标签，
#最后一行为全图设置了标题。其中使用y 轴的参数调整了标题在 y 轴方向相对于图片顶部的位置，并微微增加了默认字体的大小。
```

```python
#用于展示工作原理，并不是最好用的
fontsize=24
fig, ax = plt.subplots(2,2, figsize=(12,8))
ax[0][0].hist(df['petal width'], color='black');
ax[0][0].set_ylabel('Count', fontsize=fontsize)
ax[0][0].set_xlabel('Width', fontsize=fontsize)
ax[0][0].set_title('Iris Petal Width', fontsize=28, y=1.01)

ax[0][1].hist(df['petal length'], color='black');
ax[0][1].set_ylabel('Count', fontsize=fontsize)
ax[0][1].set_xlabel('Lenth', fontsize=fontsize)
ax[0][1].set_title('Iris Petal Lenth', fontsize=28, y=1.01)

ax[1][0].hist(df['sepal width'], color='black');
ax[1][0].set_ylabel('Count', fontsize=fontsize)
ax[1][0].set_xlabel('Width', fontsize=fontsize)
ax[1][0].set_title('Iris Sepal Width', fontsize=28, y=1.01)

ax[1][1].hist(df['sepal length'], color='black');
ax[1][1].set_ylabel('Count', fontsize=fontsize)
ax[1][1].set_xlabel('Length', fontsize=fontsize)
ax[1][1].set_title('Iris Sepal Length', fontsize=28, y=1.01)

plt.tight_layout()
#该方法将很好地自动调整子插图，以避免排版上显得过于拥挤。
```

```python
plt.close()
```

```python
#散点图
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(df['petal width'],df['petal length'], color='green')
ax.set_xlabel('Petal Width')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Scatterplot')
```

```python
plt.close()
```

```python
#线图
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(df['petal length'], color='blue')
ax.set_xlabel('Specimen Number')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Length Plot')
```

```python
#条形图
fig, ax = plt.subplots(figsize=(6,6))
bar_width = .8
labels = [x for x in df.columns if 'length' in x or 'width' in x]
ver_y = [df[df['class']=='Iris-versicolor'][x].mean() for x in labels]
vir_y = [df[df['class']=='Iris-virginica'][x].mean() for x in labels]
set_y = [df[df['class']=='Iris-setosa'][x].mean() for x in labels]
x = np.arange(len(labels))

#这里没有正确反映数量关系，代码有误
ax.bar(x, vir_y, bar_width, bottom=set_y, color='darkgrey')
ax.bar(x, set_y, bar_width, bottom=ver_y, color='white')
ax.bar(x, ver_y, bar_width, color='black')

##不想在 x 轴显示 1 到 4，因此调用了.set_xticklabels()方法并传入想要显示的列名。
#ax.set_xtickets(x+ (bar_width/2))  #可与下一行替换
ax.set_xticks(x )
ax.set_xticklabels(labels, rotation=-45, fontsize=12);
ax.set_title('Mean Feature Measurement By Class', y=1.01)
ax.legend(['Virginica','Setosa','Versicolor'])
```
