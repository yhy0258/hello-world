```python
from pandas import read_csv
from numpy import set_printoptions
```

```python
target_url=r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data=read_csv(target_url,header=None,names=['sepal length', 'sepal width','petal length', 'petal width', 'class'])
array=data.values
```

```python
set_printoptions(precision=3)
```

```python
array
```

```python
X=array[:,:-1]
y=array[:,-1]
```

# 调整数据尺度

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
transformer=MinMaxScaler(feature_range=(0,1))
newX=transformer.fit_transform(X)
```

```python
newX
```

# 正态化数据

符合高斯分布的：线性回归、逻辑回归、线性判别分析

```python
from sklearn.preprocessing import StandardScaler
```

```python
transformer=StandardScaler().fit(X)
newX=transformer.transform(X)
```

```python
newX
```

# 二值数据

01表示，在生成明确值或特征工程增加属性时使用

```python
from sklearn.preprocessing import Binarizer
```

```python
transformer=Binarizer(threshold=2.5).fit(X)
newX=transformer.transform(X)
```

# 特征工程

## 分类特征

```python
data=[
    {'price':85,'rooms':4,'neighborhood':'queen'},
    {'price':70,'rooms':3,'neighborhood':'fre'},
    {'price':65,'rooms':3,'neighborhood':'wang'},
    {'price':60,'rooms':2,'neighborhood':'fre'},
]
```

```python
from sklearn.feature_extraction import DictVectorizer
vec=DictVectorizer(sparse=False,dtype=int)
vec.fit_transform(data)
```

```python
vec.get_feature_names()
```

```python
vec=DictVectorizer(sparse=True,dtype=int)
vec.fit_transform(data)
#稀疏矩阵表示
```

## 文本特征

```python
sample=[
    'problem of evil',
    'evil queen',
    'horizon problem',
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec=CountVectorizer()
X=vec.fit_transform(sample)
X
```

```python
import pandas as pd

pd.DataFrame(X.toarray(),columns=vec.get_feature_names())
```

频繁出现的词语会有很高的权重，通过TF-IDF改善

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vec=TfidfVectorizer()
X=vec.fit_transform(sample)
pd.DataFrame(X.toarray(),columns=vec.get_feature_names())
```
