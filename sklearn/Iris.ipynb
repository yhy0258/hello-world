```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline

target_url = ("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
```

```python
df=pd.read_csv(target_url,header=None,names=['sepal length', 'sepal width','petal length', 'petal width', 'class'])
df.head()
```

```python
#
data=df[['petal length','petal width']].values
data[0:5,:]
```

```python
#
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(df['petal width'],df['petal length'], color='green')
ax.set_xlabel('Petal Width')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Scatterplot')
```

# 

```python
#
from sklearn.preprocessing import MinMaxScaler
#
from sklearn.cluster import KMeans
```

```python
transformer=MinMaxScaler(feature_range=(0,1))
newData=transformer.fit_transform(data)
```

```python
sx=newData[:,0]
sy=newData[:,1]
```

```python
#
fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(sx,sy, color='green')
ax.set_xlabel('Petal Width')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Scatterplot')
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(newData)
```

```python
#
fig, ax = plt.subplots(figsize=(6,6))
pred=KMeans(n_clusters=3,random_state=0).fit_predict(newData)
ax.scatter(sx,sy, c=pred)
ax.set_xlabel('Petal Width')
ax.set_ylabel('Petal Length')
ax.set_title('Petal Scatterplot')
```
